```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import re #Regular Expressions
import requests
```

```python

player_empty = []
position_empty = []
stat_empty = []

for page in range(1,3): #starting from page 1 to page 2
    url = "https://www.mlb.com/yankees/stats/games/2021?page={page}".format(page =page)
    print(url)
    page = requests.get(url)
    soup = bs(page.content, "html.parser")
    player_table = soup.find(class_='stats-body-table player')
    
    positions = player_table.find_all('div',class_='position-28TbwVOg')
    for position in positions:
        print(position.text.strip()) 
        position_empty.append(position.text.strip())
    
    players = player_table.find_all('a')
    for player in players:
        if player.get('aria-label') != None and player.get('aria-label') != 'MLB.com':
            print(player.get('aria-label'))
            player_empty.append((player.get('aria-label')))
   

    stats = soup.find_all('td', {'scope': 'row'}) #some stats are missing with those on the expanded-page
    for stat in stats:
        if stat.get('scope') != 'NYY':  #still printing NYY for each row
            stat_empty.append(stat.text.strip())
    
    N = 17
    sub_stat = [stat_empty[n:n+N] for n in range(0, len(stat_empty), N)]
```

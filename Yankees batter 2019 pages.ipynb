```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import re #Regular Expressions
import requests
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
player_empty = []
position_empty = []
stat_empty = []

for page in range(1,3): #starting from page 1 to page 2
    url = "https://www.mlb.com/yankees/stats/games/2019?page={page}".format(page =page)
    print(url)
    page = requests.get(url)
    soup = bs(page.content, "html.parser")
    player_table = soup.find(class_='stats-body-table player')
    
    positions = player_table.find_all('div',class_='position-28TbwVOg')
    for position in positions:
        print(position.text.strip()) 
        position_empty.append(position.text.strip())
    
    players = player_table.find_all('a')
    for player in players:
        if player.get('aria-label') != None and player.get('aria-label') != 'MLB.com':
            print(player.get('aria-label'))
            player_empty.append((player.get('aria-label')))
   

    stats = soup.find_all('td', {'scope': 'row'}) 
    for stat in stats:
        if stat.get('scope') != 'NYY': 
            stat_empty.append(stat.text.strip())
    
    N = 17
    sub_stat = [stat_empty[n:n+N] for n in range(0, len(stat_empty), N)]
```

```python
sub_stat
```

```python
hitter_df = pd.DataFrame(player_empty)
hitter_df
```

```python
position_df = pd.DataFrame(position_empty)
position_df
```

```python
sub_stat_df = pd.DataFrame(sub_stat)
sub_stat_df
```

```python
all_hitters_df = pd.concat([hitter_df,position_df],axis=1)
all_hitters_df
```

```python
hitters_stats = pd.concat([all_hitters_df, sub_stat_df],axis=1)
hitters_stats
```
